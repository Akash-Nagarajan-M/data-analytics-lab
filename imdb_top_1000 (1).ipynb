## updated regression code


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df2=df[['Gross','Runtime','No_of_Votes']]
df2['Gross']/=1000000
df2[['Gross','Runtime','No_of_Votes']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df2, df['IMDB_Rating'], test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
df2.info()
```

```python
svr = SVR(kernel='linear')
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print('Mean squared error: ', mse)
print('R-squared: ', r2)
```

## Naive Bayes classification ,i guess :)


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv('Breast_cancer_data.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['diagnosis'], axis=1), df['diagnosis'], test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=0)
recall = recall_score(y_test, y_pred, pos_label=0)
f1 = f1_score(y_test, y_pred, pos_label=0)
```

```python
print('Accuracy: ', accuracy)
print('Precision: ', precision)
print('Recall: ', recall)
print('F1-score: ', f1)
```
